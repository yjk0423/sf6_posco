### 시각화 라이브러리 사용 - matplotlib

```python
import matplotlib.pyplot as plt

plt.title("Basic Graph")
# plt.plot([1,2,3,4]) # 리스트가 하나가 들어가면 Y축임
plt.plot([1,2,3,4],[10,40,20,30]) #가로축 x : 1,2,3,4 세로축 y : 10,40,20,30 (1, 10) , (2, 40)
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.rc("font", family="Malgun Gothic") # 한글 오류를 해결하기 위해 사용함

plt.title("시험 성적 그래프")
plt.plot([1,2,3,4],[90,80,85,70]) #가로축 x : 1,2,3,4 세로축 y : 10,40,20,30 (1, 10) , (2, 40)
plt.xlabel("학생번호")
plt.ylabel("시험점수")
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.rc("font", family="Malgun Gothic") # 한글 오류를 해결하기 위해 사용함

plt.title("범례 및 색상 표시")
plt.plot([10,20,30,40], label="비만율", color='red')
plt.plot([40,30,20,10], label="출생율", color='g')
plt.legend() # 바로위의 label을 사용을 하면 필수적으로 작성해야 출력이 됨 (그래프의 소제목)


plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc("font", family="Malgun Gothic") # 한글 오류를 해결하기 위해 사용함

plt.title("마커 표시")
plt.plot([10,20,30,40], 'b.', label="circle")
plt.plot([40,30,20,10], 'r^', label="triangle")
plt.legend() # 바로위의 label을 사용을 하면 필수적으로 작성해야 출력이 됨 (그래프의 소제목)


plt.show()
```
