#### hist(데이터리스트)

```python
import matplotlib.pyplot as plt
import random

# plt.hist([1,1,2,3,4,5,5,5,5,7])
# plt.show()


dice = []

for i in range(10):
    dice.append(random.randint(1,6))

print(dice)

plt.hist(dice, bins=6) # 가로축의 구간의 개수
plt.show()
```

```python
import csv
import math
import matplotlib.pyplot as plt
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)

    next(data)

    temp_high = []
    for row in data:
        if row[-1] != "":
            row[0] = row[0].replace("\t","")
            # if int(row[0].split("-")[0]) >= 1980:
            if row[0].split("-")[1] == "08":
                temp_high.append(float(row[-1]))
                
            
# print(max(temp_high))

plt.rc("font", family="Malgun Gothic")
plt.title("1980년 이후 서울의 최고 기온")
plt.hist(temp_high,bins = 100, color= "r")
plt.xlabel("온도")
plt.ylabel("빈도수")
plt.show()
```
